# Numpy
NumPy is a Python library used for working with arrays.

It also has functions for working in domain of linear algebra, fourier transform, and matrices.

NumPy was created in 2005 by Travis Oliphant. It is an open source project and you can use it freely.

NumPy stands for Numerical Python.

# Dot Product Example (Nut Butter Sales)

```python
np.random.seed(0)
# No of Jar Sold
sales_amt = np.random.randint(20, size=(5, 3))
sales_amt
```

```python
# Create Weekly sales dataframe
weekly_sales = pd.DataFrame(sales_amt,index=["mon","tue","wed","thu","fri"],columns=["Almond Butter","Peanut Butter","Cashew Butter"])
weekly_sales
```

```python
# Create Prices Array

prices = np.array([10, 8, 12])
butter_prices = pd.DataFrame(prices.reshape(1,3),index=["Price"],columns=["Almond Butter","Peanut Butter","Cashew Butter"])
```

```python
butter_prices
```

```python
prices.dot(sales_amt)
```

```python
# Shapes arn't aligned lets transpose   

total_sales = prices.dot(sales_amt.T)
total_sales
```

```python
# Create Daily Sales

butter_prices
```

```python
weekly_sales
```

```python
butter_prices.shape, weekly_sales.shape
```

```python
daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```

```python
weekly_sales["Total $"] = daily_sales.T
daily_sales.T
```

```python
weekly_sales
```
