# Introduction to Matplotlib

Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

Matplotlib was created by John D. Hunter.

Matplotlib is open source and we can use it freely.

Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.

# Matplotlib Example Workflow

```python
# Prepeare Data
x = np.arange(0,100,5)
y= np.arange(0,100,5)

# Setup Plot
fig, ax = plt.subplots(figsize=(10, 10))  # (width,height)

# Plot Data
ax.plot(x, y)

# Customize Plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

# Save and show the whole figure
fig.savefig("./images/sample_plot.png")
```

# Two Option for subplots

```python
# Subplot option 1

fig,((ax1,ax2),(ax3,ax4)) =plt.subplots(figsize=(10,5),nrows=2,ncols=2)

# Plot to each different axis
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(list(nut_buuter_prices.keys()),list(nut_buuter_prices.values()))
ax4.hist(x);
plt.show()
```

```python
# subplot option 2

fig,ax = plt.subplots(nrows=2,ncols=2 ,figsize=(10,5))

# Plot to different axes

ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(list(nut_buuter_prices.keys()),list(nut_buuter_prices.values()))
ax[1,1].hist(x);
```

# Heart Disease 

```python
heart_disease =  pd.read_csv("./data/heart-disease.csv")
```

```python
heart_disease.head(2)
```

```python
# Historgram of age column

heart_disease["age"].plot.hist();
```

```python
heart_disease.plot.hist(subplots=True,figsize = (10,30));
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
```

```python
over_50
```

```python
# Pyplot Methpd
over_50.plot(x="age",y="chol",kind="scatter",c ="target");
```

```python
# OO Method
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(x="age",y="chol",kind="scatter",c ="target",ax=ax);
ax.set_xlim([45,100])
```

```python
fig,ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
ax.set(title="Heart Disease and Cholestrol Level",xlabel="Age", ylabel="Cholestrol")
ax.legend(*scatter.legend_elements(),title="Target")

# adding horizontal line

ax.axhline(over_50["chol"].mean(),linestyle="--")

plt.show()
```

# Subplot of chol,age,thalach



```python
fig,(ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize =(10,10),sharex=True)

sc1 = ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
ax0.axhline(over_50["chol"].mean(),ls="--")
ax0.set(title="Heart Disease and Cholestrol Levels",ylabel="chol")
ax0.legend(*sc1.legend_elements(),title="Target")


sc2 = ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
ax1.axhline(over_50["thalach"].mean(),ls="--")
ax1.set(title="Heart Disease and Max Heart Rate", xlabel="Age",ylabel="Thalach")
ax1.legend(*sc2.legend_elements(),title="Target")

fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold")

```

```python
# Set the style

plt.style.use("seaborn-v0_8-whitegrid")

fig,ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter")
ax.set(title="Heart Disease and Cholestrol Level",xlabel="Age", ylabel="Cholestrol")
ax.legend(*scatter.legend_elements(),title="Target")

# adding horizontal line

ax.axhline(over_50["chol"].mean(),linestyle="--")

plt.show()
```

# Customizing the x and y axis

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

sc1 = ax0.scatter(x=over_50["age"], y=over_50["chol"],
                  c=over_50["target"], cmap="winter")
ax0.set_xlim([50,80])
ax0.axhline(over_50["chol"].mean(), ls="--")
ax0.set(title="Heart Disease and Cholestrol Levels", ylabel="Cholestrol")
ax0.legend(*sc1.legend_elements(), title="Target")


sc2 = ax1.scatter(x=over_50["age"], y=over_50["thalach"],
                  c=over_50["target"], cmap="winter")
ax1.axhline(over_50["thalach"].mean(), ls="--")
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age", ylabel="Max Heart Rate")
ax1.legend(*sc2.legend_elements(), title="Target")
ax1.set_xlim([50,80])
ax1.set_ylim([50,200])

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
fig.savefig('./images/heart_diesase_analysis')
```
