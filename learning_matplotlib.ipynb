# Introduction to Matplotlib

Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

Matplotlib was created by John D. Hunter.

Matplotlib is open source and we can use it freely.

Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.

# Matplotlib Example Workflow

```python
# Prepeare Data
x = np.arange(0,100,5)
y= np.arange(0,100,5)

# Setup Plot
fig, ax = plt.subplots(figsize=(10, 10))  # (width,height)

# Plot Data
ax.plot(x, y)

# Customize Plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

# Save and show the whole figure
fig.savefig("./images/sample_plot.png")
```

```python
# Make Plot from dictionary

nut_buuter_prices = {
    "Almond Butter": 10,
    "Peanut Butter": 5,
    "Cashew Butter": 8
}
```

```python
fig,ax = plt.subplots(figsize=(8,5))
ax.bar(nut_buuter_prices.keys(),nut_buuter_prices.values())
ax.set(title="Nut Butter Prices",ylabel="Prices $")
ax.set_xlabel("Types of Nut Butter")
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(8,5))
ax.barh(list(nut_buuter_prices.keys()),list(nut_buuter_prices.values()))
ax.set(title="Nut Butter Prices",ylabel="Types of Nut Butter")
ax.set_xlabel("Prices $")
plt.show()
```

# Two Option for subplots

```python
# Subplot option 1

fig,((ax1,ax2),(ax3,ax4)) =plt.subplots(figsize=(10,5),nrows=2,ncols=2)

# Plot to each different axis
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(list(nut_buuter_prices.keys()),list(nut_buuter_prices.values()))
ax4.hist(x);
plt.show()
```

```python
# subplot option 2

fig,ax = plt.subplots(nrows=2,ncols=2 ,figsize=(10,5))

# Plot to different axes

ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(list(nut_buuter_prices.keys()),list(nut_buuter_prices.values()))
ax[1,1].hist(x);
```
